* Those columns referes to IDs that should be object instead of fload and int data types [`tweet_id`,`in_reply_to_status_id`,`in_reply_to_user_id`,`retweeted_status_id`,`retweeted_status_user_id`].

```python
# changing some columns data types.

Twitter_Archive=Twitter_Archive.astype(dtype={'tweet_id':'object','in_reply_to_status_id':'object','in_reply_to_user_id':'object','retweeted_status_id':'object','retweeted_status_user_id':'object'},copy=True)
Twitter_Archive.info()
```

* `None` values should be replaced with `NaN` (this may help us in cleaning and merging and save us some few lines of code).


```python
Twitter_Archive.iloc[:,13:].astype(str).replace({'None':np.nan},inplace=True)
Twitter_Archive.iloc[:,13:]
```

```python
Twitter_Archive.iloc[:,13:].info()
```

```python
# since the above code doesn't work we will have to fill each column separately
# posted a question for help on Stackoverflow
#https://stackoverflow.com/questions/63405869/replace-didnt-work-properly-for-dataframe
Twitter_Archive["doggo"].replace({'None':np.nan},inplace=True)
Twitter_Archive["floofer"].replace({'None':np.nan},inplace=True)
Twitter_Archive["pupper"].replace({'None':np.nan},inplace=True)
Twitter_Archive["puppo"].replace({'None':np.nan},inplace=True)
Twitter_Archive.iloc[:,13:]
```

```python
Twitter_Archive.info()
```

* [`timestamp`,`retweeted_status_timestamp`] should be conferted to datatime data type.

```python
#this code doesn't work
#Twitter_Archive.loc[:,["retweeted_status_timestamp","timestamp"]] = pd.to_datetime(Twitter_Archive.loc[:,["retweeted_status_timestamp","timestamp"]])
```

```python
# since the above code doesn't work we will have to convert each column one at a time.
Twitter_Archive["retweeted_status_timestamp"] = pd.to_datetime(Twitter_Archive["retweeted_status_timestamp"])
Twitter_Archive["timestamp"] = pd.to_datetime(Twitter_Archive["timestamp"])
Twitter_Archive.info()
```

* There are some missing `expanded_urls` values they should be filled.

```python
url = "https://twitter.com/dog_rates/status/"
L = Twitter_Archive[ Twitter_Archive["expanded_urls"].isnull()].index
for x in L:
    Twitter_Archive.iloc[x,9] = url + str(Twitter_Archive.iloc[x,0])
Twitter_Archive.info()
```

```python
Twitter_Archive.iloc[L,[0,9]]
```

* `name` contain some`None` values which should be replaced with `NaN`

```python
Twitter_Archive["name"].replace({'None':np.nan},inplace=True)
Twitter_Archive["name"].value_counts()
```
